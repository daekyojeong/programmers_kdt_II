## I. Matplotlib 시작하기

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
# inline을 통해 활성화
# 꺽은선 그래프
plt.plot([2,4,2,4,2]) # 인자는 y, x는 인덱스
plt.show() # plt 확인
```

```python
plt.figure(figsize=(3,3)) # plotting을 할 도면을 선언, 

plt.plot([1,2,3,4,5]) # 인자는 y, x는 인덱스
plt.show() # plt 확인
```

```python
x = np.array([1,2,3,4,5]) #정의역
y = np.array([1,4,9,16,25]) #f(x)

plt.plot(x,y)
plt.show()
```

```python
# np.arange(a,b,c)  c: 극소의 값

x = np.arange(-10,10,0.01)
plt.xlabel("x value")
plt.ylabel("f(x) value")
plt.plot(x, x**2)
plt.show()

```

## II. Matplotlib Case Study

```python
# x, y축 범위 설정

x = np.arange(-10,10,0.01)
plt.xlabel("x value")
plt.ylabel("f(x) value")

plt.axis([-5,5,0,25]) # [x_min, x_max, y_min, y_max]
plt.plot(x, x**2)
plt.show()
```

```python
#x, y축 눈금 설정

x = np.arange(-10,10,0.01)
plt.xlabel("x value")
plt.ylabel("f(x) value")

plt.axis([-5,5,0,25]) # [x_min, x_max, y_min, y_max]

plt.xticks([i for i in range(-5,6,1)])
plt.yticks([i for i in range(0,27,3)])

plt.title("y = x^2")

plt.plot(x, x**2, label='trend')

plt.legend()

plt.show()
```

## III. Matplotlib Case Study

```python
# 꺾은선 그래프 (plot)
# 시계열 데이터 등 

x = np.arange(20)
y = np.random.randint(0,20,20)
plt.axis([0,20,0,20]) # [x_min, x_max, y_min, y_max]
plt.yticks([i for i in range(0,21,5)])

plt.plot(x,y)
plt.show()

# Extra : y축 20까지 5단위로
```

```python
# 산점도 (Scatter Plot)
# 두 축사이의 상관관계를 찾기위해

plt.scatter(x,y)
plt.show()
```

```python
# 박스 그림 (box plot)

# 수치형 데이터에 대한 정보 (Q1, Q2, Q3, min, max)
# 상위 25% 50% 75% 데이터
plt.title("Box plot of y")
plt.boxplot(y)
plt.show()
```

```python
plt.title("Box plot of y")
plt.boxplot((x,y))
plt.show()
```

```python
# 막대그래프 (Bar plot)
# 범주형데이터
plt.xticks(np.arange(0,20,1))
plt.bar(x,y)
plt.show()
```

```python
# 히스토그램
# "계급"으로 나타냄 : 0,1,2가 아니고 0~2까지 범주형 데이터로 그림
plt.hist(y, bins=np.arange(0,20,2))
plt.xticks(np.arange(0,20,2))
plt.show()
```

```python
# 원형 그래프 (pie chart)
# 비율 확인에 용이
z = [100,300,200,400]
plt.pie(z, labels=['one','two','three','four'])
plt.show()
```

## IV. The 멋진 그래프, Seaborn Case Study

```python
# 커널밀도그림
# 카운트그림
# 캣그림
# 스트립그림
# 히트맵

import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
x = np.arange(0,22,2)
y = np.random.randint(0,20,20)

plt.hist(y,bins=x)
plt.show()
```

```python
sns.kdeplot(y)
plt.show()
```

```python
sns.kdeplot(y, shade=True)
plt.show()
```

```python
vote_df = pd.DataFrame({'name':['andy','bob','cat'], 'vote':[True,True,False]})
vote_count = vote_df.groupby('vote').count()
```

```python
plt.bar(x=[False, True], height=vote_count['name'])
plt.show()
```

```python
sns.countplot(x = vote_df['vote'])
plt.show()
```

```python
covid = pd.read_csv("./country_wise_latest.csv")
```

```python
s = sns.catplot(x='WHO Region',y='Confirmed', data=covid, kind = 'violin') #kind = '' (default = strip)
s.fig.set_size_inches(10,6)
plt.show()
```

```python
sns.stripplot(x='WHO Region',y='Recovered',data=covid)
plt.show()
```

```python
sns.swarmplot(x='WHO Region',y='Recovered',data=covid)
plt.show()
```

```python
covid.corr()
```

```python
sns.heatmap(covid.corr())
plt.show()
```
